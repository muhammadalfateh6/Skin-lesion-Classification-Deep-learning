```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Training = [93.25 , 94.8, 97.48, 97.43 , 98.7]
Testing = [91.26 , 92.5, 95.7, 96.2, 95.1]
w = 0.4
labels = ['VGG16', 'InceptionV3', 'ResNet50','Xception','VGG19']
colorss=['green', 'red','blue', 'gold', 'peru']
x = np.arange(len(labels))
x1 = [i+w for i in x]

plt.bar(x, Training, w, label='Training Accuracy')
plt.bar(x1, Testing, w, label='Testing Accuracy')

plt.xlabel('Number Of Images')
plt.ylabel('Number Of Images')
plt.legend()
plt.xticks(x, labels)
plt.title('Class Distribution In ISIC2018 Dataset')
```
