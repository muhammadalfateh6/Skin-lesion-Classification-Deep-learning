```
# Muhammad Alfateh's thesis work
# implermanting on 4th semester MSAI 
import tensorflow as tf
from tensorflow.keras.callbacks import Callback
from tensorflow.keras.preprocessing.image import ImageDataGenerator

vgg_model = tf.keras.applications.vgg19.VGG19(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

for layer in vgg_model.layers:
    layer.trainable = False

main_directory = '/content/drive/MyDrive/PH2 Ready'
train_directory = '/content/drive/MyDrive/PH2 Ready'
test_directory = '/content/drive/MyDrive/Test ph2'

train_path = '/content/drive/MyDrive/PH2 Ready'
class_1_train_path = '/content/drive/MyDrive/PH2 Ready/Common'
class_2_train_path = '/content/drive/MyDrive/PH2 Ready/Atypical'
class_3_train_path = '/content/drive/MyDrive/PH2 Ready/Melanoma'

test_path = '/content/drive/MyDrive/Test ph2'
class_1_test_path = '/content/drive/MyDrive/Test ph2/Common'
class_2_test_path = '/content/drive/MyDrive/Test ph2/Atypical'
class_3_test_path = '/content/drive/MyDrive/Test ph2/Melanoma'

data_generator = ImageDataGenerator(rescale=1./255)

train_data = data_generator.flow_from_directory(train_directory,
                                                 target_size=(224, 224),
                                                 batch_size=32,
                                                 class_mode='categorical',
                                                 shuffle=True,
                                                 save_to_dir=train_path,
                                                 save_format='bmp')

test_data = data_generator.flow_from_directory(test_directory,
                                                target_size=(224, 224),
                                                batch_size=32,
                                                class_mode='categorical',
                                                shuffle=False,
                                                save_to_dir=test_path,
                                                save_format='bmp')
class calls(Callback):
    def on_train_begin(self, logs=None):
        print("Hey Muhammad, the training started")
    def on_epoch_begin(self, epoch, logs=None):
        print(f"Strating epoch number {epoch}")
    def on_epoch_end(self, epoch, logs=None):
        print(f"Epoch number {epoch} ended")              
    def on_train_end(self, logs=None):
        print("Muhammad, Training ended for good")

calling =[calls()]

model = tf.keras.models.Sequential([
        vgg_model,
        tf.keras.layers.Conv2D(64, (3, 3), padding='same', activation='relu'),
        tf.keras.layers.MaxPooling2D((2, 2), padding='same'),
        tf.keras.layers.Conv2D(128, (3, 3), padding='same', activation='relu'),
        tf.keras.layers.MaxPooling2D((2, 2), padding='same'),
        tf.keras.layers.Conv2D(256, (3, 3), padding='same', activation='relu'),
        tf.keras.layers.MaxPooling2D((2, 2), padding='same'),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(256, activation='relu'),
        tf.keras.layers.Dropout(0.5),
        tf.keras.layers.Dense(3, activation='softmax')
])

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```
history = model.fit(train_data, callbacks = calling, epochs=50 , verbose=1)
```

```
# Evaluate the model on the test set
test_loss, test_acc = model.evaluate(test_data)
```
